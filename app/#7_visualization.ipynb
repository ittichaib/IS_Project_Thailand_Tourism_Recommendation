# Data Visualization for Tourism Recommendation System

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('../dataset/reviews/TripAdvisor_reviews_EN.csv')

# Distribution of Target Variable
sns.countplot(data['rating'])
plt.title('Distribution of Target Variable')
plt.show()
```

```python
data
```

```python
# Feature Correlation Heatmap
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# PCA Visualization
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data.drop('rating', axis=1))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=data['rating'], cmap='viridis')
plt.title('PCA of Features')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```
